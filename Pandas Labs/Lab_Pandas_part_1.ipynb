### Cleaning and wrangling data

- Some car brand names refer to the same brand. Replace all brand names that contain the word "Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use `loc` with boolean indexing.

- Convert CO2 Emissions from Grams/Mile to Grams/Km

- Create a binary column that solely indicates if the transmission of a car is automatic or manual. Use `pandas.Series.str.startswith` and .

- convert MPG columns to km_per_liter

Note:
<br>Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters



```python
#1
df.loc[df["Make"].str.contains("Dutton",case=False),"Make"] = "Dutton"

df.loc[df["Make"].str.contains("Dutton",case=False)]
```

```python
#2

#df["CO2 Emission Grams/km"]=df["CO2 Emission Grams/Mile"]/1.60934
#df=df.drop("CO2 Emission Grams/Mile",axis=1)

df["CO2 Emission Grams/Mile"] /= 1.60934
df.rename(columns={"CO2 Emission Grams/Mile": "CO2 Emission Grams/km"})
df
```

```python
#3


#df["Transmission"]=list(map(lambda x: x.bool() if df.loc[df["Transmission"].str.startswith("Auto"),"Transmission"] else x.bool,df["Transmission"]))


#bool_list=list(map(lambda x: True if df.loc[df["Transmission"]].str.startswith("Auto") else False,df["Transmission"]))

#booli=df.loc[df["Transmission"]].str.startswith("Auto",na=False)

#bool_list=df["Transmission"].str.startswith("Auto",na=False)

def pleasework(element):
    if element==True:
        return "Automatic"
    else:
        return "Manual"

df["Auto/Manual"]=df["Transmission"].apply(lambda x: "Auto" if x.startswith("Auto") else "Manual")
#df["Auto/Manual"]=
#df["Auto/Manual"]=df["Auto/Manual"].apply(pleasework)
df
```

```python
#4

#Converting the three MPG (Miles per Gallon) columns to kilometers per liter

df["City MPG"] = df["City MPG"].apply(lambda x: x*1.60934/3.78541)

df["Highway MPG"] = df["Highway MPG"].apply(lambda x: x*1.60934/3.78541)

df["Combined MPG"] = df["Combined MPG"].apply(lambda x: x*1.60934/3.78541)

df=df.rename(columns={"City MPG":"City KPL","Highway MPG":"Highway KPL","Combined MPG":"Combined KPL"})

df
```
